```python
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
import pandas as pd
import numpy as np
```

```python
##Read in training data##
data=pd.read_csv('task1_2021.training.csv')
data.head()
```

```python
##Selecting variables##
a=['f0','f1','f2','f3','f4','f5','f6','f7','f8','f9','f10','f11','target']
a.remove("target")
features=a
features
lr_xx=data[features].to_numpy()
lr_xx[0:5]
```

```python
##Dependent variable##
lr_yy=data['target']
lr_yy[0:5]
```

```python
##Read in test data##
data2=pd.read_csv('task1_2021.test.csv')
data2.head()
```

```python
##Independent variables##
lr_xx_test=data2.to_numpy()
lr_xx_test[0:5]
```

```python
# Building Models##
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(lr_xx, lr_yy)

# Make predictions using the testing set
lr_yy_pred = regr.predict(lr_xx_test)

# The coefficients
print('Coefficients: \n', regr.coef_)

# intercept
print('Intercept: \n', regr.intercept_)
```

```python
##Making output file##
ID=[]
for i in range(2000):
    ID.append(i+1)

predicted=list(lr_yy_pred)

dic={'id':ID , 'Predicted':predicted}
output=pd.DataFrame(dic)
output.to_csv('task1.csv',index=False)
```
