```python
import pandas as pd
import numpy as np

data=pd.read_csv('task3_2021.training.csv')
data['target'].value_counts()

##Selecting variables##
target=data['target']
training=data.loc[:,data.columns!='target']
training.head()
```

```python
##Read in Testing data##
data2=pd.read_csv('task3_2021.test.csv')
data2.head()
```

```python
##Label Encoding##
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
training_lb=pd.DataFrame(training)
training_lb['f15']=lb.fit_transform(training_lb['f15'])
training_lb['f16']=lb.fit_transform(training_lb['f16'])
training_lb['f17']=lb.fit_transform(training_lb['f17'])
training_lb.head()
```

```python
##Label Encoding(test)##
lb=LabelEncoder()
test_lb=pd.DataFrame(data2)
test_lb['f15']=lb.fit_transform(test_lb['f15'])
test_lb['f16']=lb.fit_transform(test_lb['f16'])
test_lb['f17']=lb.fit_transform(test_lb['f17'])
test_lb.head()
```

```python
nor = pd.concat([training_lb,test_lb])
from sklearn import preprocessing
test2 = nor.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(test2)
nor = pd.DataFrame(x_scaled)
nor
```

```python
training_lb = nor.iloc[0:10000,:]
test_lb = nor.iloc[10000:,:]
```

```python
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
# load dataset

dataset = training_lb.values
#min_max_scaler = preprocessing.MinMaxScaler()
X = dataset[:,0:19].astype(float)
#X = dataset[:,0:11].astype(float)
Y = target.to_numpy()
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(encoded_Y)
dummy_y
```

```python
def baseline_model():
    model = Sequential()
    model.add(Dense(175, input_dim=19, activation='relu'))
    model.add(Dense(175, activation='relu'))
    #model.add(Dense(115, activation='relu'))
    model.add(Dense(10, activation='softmax'))
	# Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
estimator = KerasClassifier(build_fn=baseline_model, epochs=100, batch_size=50, verbose=0)
kfold = KFold(n_splits=10, shuffle=True)
results = cross_val_score(estimator, X, dummy_y, cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```python
estimator.fit(X, Y)
dic = {'Id': range(1,2001), 'Predicted': estimator.predict(test_lb)}
output = pd.DataFrame(dic)
output.to_csv("task3layer2epoch100.csv", index=False)
```
